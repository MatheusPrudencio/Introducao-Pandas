```python
import pandas as pd

%matplotlib inline
```

```python
#Ler arquivo xlsx (excel)
#skiprows pula as 2 primeiras linhas
df = pd.read_excel('2120_0.xlsx', skiprows=2)

df.sample(6)
```

```python
#skipfooter exclui as ultimas linhas (importante pois isso acontece algumas vezes nas bases de dados) que estao ruins
df = pd.read_excel('2120_0.xlsx', skiprows=2, skipfooter=16)
```

```python
df.tail(6)
```

```python
df.dtypes
```

```python
# Converter o campo Data de lançamento de object (str) para data

df['Data de lançamento'] = pd.to_datetime(df['Data de lançamento'])
```

```python
# Gracas ao erro anterior sabemos que algum campo nao esta com o formato str

df['Data de lançamento'].value_counts()
```

```python
#Como data = relancamento a funcao nao funciona
#Exibir algumas dessas p/ analise

df[df['Data de lançamento'] == 'Relançamento'].head(6)
```

##Decisao sobre dados incorretos/faltantes/divergentes

```python
df.shape
```

```python
# Criar um novo DataFrame sem os relancamentos

df_new = df[df['Data de lançamento'] != 'Relançamento'].copy()
```

```python
# Converter as Datas

df_new['Data de lançamento'] = pd.to_datetime(df_new['Data de lançamento'])
```

```python
df_new.dtypes
```

```python
# Quais as datas de lancamentos usando o tipo datetime?

df_new['Data de lançamento'].dt.year.value_counts()
```
